```python
# Using heatmap to visualize a correclation matrix in python
import numpy as np
import matplotlib.pyplot as plt

# Example data: Randomly generated correlation matrix
data = np.random.random((10, 10))
correlation_matrix = np.corrcoef(data)
plt.imshow(correlation_matrix, cmap=plt.cm.hot, interpolation='nearest')
plt.colorbar()
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# A 2D Matrix HeatMap
# Generate a random 2D matrix
rng = np.random.default_rng(27446968)
matrix = rng.random((10, 10))

# Plot the 2D matrix
plt.figure(figsize=(5, 6))
plt.imshow(matrix, cmap="viridis", interpolation='nearest')
plt.colorbar()
plt.title('2D Matrix Heatmap')
plt.show()
```

```python
# Cosine function as a function of time

# Generate time values
time = np.linspace(0, 2 * np.pi, 1000)

# Compute the cosine of time
cosine_values = np.sin(time)

plt.figure(figsize=(10, 4))
plt.plot(time, cosine_values, label='cos(t)')
plt.title('Cosine Function')
plt.xlabel('Time')
plt.ylabel('cos(t)')
plt.legend()
plt.grid(True)
plt.show()
```
